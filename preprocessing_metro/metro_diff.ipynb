```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rc('font',family='AppleGothic')
```

```python
df=pd.read_csv('../raw_data/seoul_metro_raw/merged_raw_metro.csv',index_col=0)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df_raw = df.copy()
```

```python
temp=df[df['구분']=='하차']
```

```python
temp.drop('구분',axis=1,inplace=True)
```

```python
from itertools import cycle, islice
my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, 10))
```

```python
temp_2017=temp[temp.index.year == 2017]

before_noon=['05~06', '06~07', '07~08', '08~09', '09~10', '10~11', '11~12']
after_noon = ['12~13', '13~14', '14~15', '15~16', '16~17']
dinner =['17~18', '18~19', '19~20','20~21', '21~22', '22~23', '23~24', '00~01']

temp.loc[:,before_noon].sum(axis=1)

merged_temp = temp.copy()
merged_temp=pd.DataFrame(merged_temp.iloc[:,0])

merged_temp['before_noon'] = temp.loc[:,before_noon].sum(axis=1)
merged_temp['after_noon'] = temp.loc[:,after_noon].sum(axis=1)
merged_temp['dinner'] = temp.loc[:,dinner].sum(axis=1)
merged_temp['total'] = merged_temp.iloc[:,1:].sum(axis=1)

tmp=merged_temp[merged_temp.index.year == 2017]

a=(tmp.groupby('역명').max()-tmp.groupby('역명').mean()) / tmp.groupby('역명').std()
a.sort_values('total',ascending=False)
plt.figure(figsize=(14,7))

plt.subplot(2,2,1)
a['before_noon'].sort_values().head(10).plot(kind='bar',color=my_colors)
plt.xticks(rotation=25)
plt.title('before_noon',fontsize=15)
plt.subplot(2,2,2)
a['after_noon'].sort_values().head(10).plot(kind='bar',color=my_colors)
plt.xticks(rotation=25)
plt.title('after_noon',fontsize=15)
plt.subplot(2,2,3)
a['dinner'].sort_values().head(10).plot(kind='bar',color=my_colors)
plt.xticks(rotation=25)
plt.title('dinner',fontsize=15)
plt.subplot(2,2,4)
a['total'].sort_values().head(10).plot(kind='bar',color=my_colors)
plt.title('total',fontsize=15)
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```
